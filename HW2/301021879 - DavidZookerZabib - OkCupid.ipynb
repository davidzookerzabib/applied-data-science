### Tidying up - Translating the offspring responses into meaningful data

** View Offsrpings Responses **

```python
NumOfKids = df.groupby('offspring')
NumOfKids.head()
```

** Analyze and Categorize Offsrpings Responses **

```python
def HaveOffsprings(x):
    if (type(x) == float):
        return None
    if 'has a kid' in x: #"a kid" means one kid
        return 'Yes, One Kid'
    elif 'has kids' in x: #Plurals
        return 'Yes, More Than One'
    else: #No kids
        return 'No'
    
def WantOffsprings(x):
    if (type(x) == float):
        return None
    if 'want' in x: #want indicator
        if 'doesn&rsquo;t want' in x: #doesn't want
            return 'No'
        else: #want
            return 'Yes'
    elif 'wants' in x: #Also want
        return 'Yes'
    else: #does not want
        return 'No'
    
HaveOffsprings = df.offspring.apply(lambda x: HaveOffsprings(x))
WantOffsprings = df.offspring.apply(lambda x: WantOffsprings(x))
df['HaveOffsprings'] = HaveOffsprings
df['WantOffsprings'] = WantOffsprings
```

** Have Offsprings vs Income **

```python
sns.barplot(x='HaveOffsprings',y='income', hue='sex',data=df)
```

** Want Offsprings vs Income **

```python
sns.barplot(x='WantOffsprings',y='income', hue='sex',data=df)
```

** Offsprings Rank **
 <br />
I'll rank the users as follows: <br />
ND: No offsprings and Don't want any <br />
NW: No offsprings and Want <br />
OD: One offspring and Don't want more <br />
OW: One offspring and Want more <br />
MD: Multiple offsprings and Don't want more <br />
MW: Multiple offsprings and Want more <br />

```python
def RankOffsprings(x):
    have = x['HaveOffsprings']
    want = x['WantOffsprings']
    if (have == None):
        return None
    if 'Yes' in have: #Have kids
        if 'More' in have: #More than one
            if 'Yes' in want: #Want more
                return 'MW'
            else: #Does not want more
                return 'MD'
        else: #One kid
            if 'Yes' in want: #Want more
                return 'OW'
            else: #Does not want more
                return 'OD'
    else: #No kids
        if 'Yes' in want: #Wants
            return 'NW'
        else: #Does not want
            return 'ND'
        
RankOffsprings = df.apply(lambda x: RankOffsprings(x), axis=1)
df['RankOffsprings'] = RankOffsprings
```

** Offsprings Rank vs Income **

```python
sns.barplot(x='RankOffsprings',y='income', hue='sex',data=df)
```

** Number of Users according to Gender **

```python
def GroupSumSort(df,col):
    Group = df.groupby(col) #Group df by specific column
    GroupSum = Group[col].count() #Count the number of occourences
    return GroupSum.sort_values(ascending=False) #Sort descending

GenderSort = GroupSumSort(df,'sex')
GenderSort.head()
```

** Number of Users according to Orientation **

```python
OrientSort = GroupSumSort(df,'orientation')
OrientSort.head()
```

** Sex and Sexual Orientation **

```python
ax = sns.barplot(x='orientation', y='orientation', orient='v', hue='sex', data=df, estimator=lambda x: len(x) / len(df) * 100)
ax.set(ylabel="Percentage")
```

** Number of Users vs Age **

```python
GrAge = df.groupby('age').count()
GrAge.plot(y='sex', legend=None)
plt.ylabel('Number of Users')
```

** Age and Sexual Orientation **

```python
GrAgeOrient = df.groupby(['age','orientation']).count()
AgeOrient = GrAgeOrient.sex.unstack('orientation')
plt.figure(figsize=(10, 16))
sns.heatmap(AgeOrient.fillna(0), annot=True)
```

** Age and Gender **

```python
GrAgeSex = df.groupby(['age','sex']).count()
AgeSex = GrAgeSex.status.unstack('sex')
plt.figure(figsize=(10, 16))
sns.heatmap(AgeSex.fillna(0), annot=True)
```

** Age and Mean Height **

```python
GrAgeMale = df.loc[df['sex'] == 'm'].groupby(['age']) #Group all males by age
AgeMaleHeight = GrAgeMale.height.mean() # Calculate their mean height
GrAgeFemale = df.loc[df['sex'] == 'f'].groupby(['age'])
AgeFemaleHeight = GrAgeFemale.height.mean()

plt.plot(AgeMaleHeight,'r', label='Male Mean Height') 
plt.plot(AgeFemaleHeight,'b', label='Female Mean Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
